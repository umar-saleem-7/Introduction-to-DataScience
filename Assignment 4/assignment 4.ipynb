```python
import numpy as np
```

```python
# Task 1
locations = 500
days = 365

temperature = np.random.uniform(-10, 40, (locations, days))

humidity = np.random.uniform(0, 100, (locations, days))

print("Temperature: \n",temperature)
print("Humidity: \n",humidity)
```

```python
# Task 2
total = temperature.size
missing_count = int(total * 0.05)

temperature_with_missing_data = temperature.copy()
humidity_with_missing_data = humidity.copy()

temperature_missing_indice = np.random.choice(total, missing_count, replace=False)
humidity_missing_indices = np.random.choice(total, missing_count, replace=False)

np.put(temperature_with_missing_data, temperature_missing_indice, None)
np.put(humidity_with_missing_data, humidity_missing_indices, None)

temperature_missing_count = np.isnan(temperature_with_missing_data).sum()
humidity_missing_count = np.isnan(humidity_with_missing_data).sum()

print("Temperature missing data count: ", temperature_missing_count)
print("Himdity missing data count: ", humidity_missing_count)
print("Total missing data: ", temperature_missing_count + humidity_missing_count)
```

```python
# Task 3
temperature_fahrenheit = temperature_with_missing_data * 9/5 + 32

# A common formula for "feels like" index is: DI = T - (0.55 - 0.0055 * RH) * (T - 58),

# Calculate discomfort index while handling NaN values (ignoring NaNs)
discomfort_index = temperature_fahrenheit - (0.55 - 0.0055 * humidity_with_missing_data) * (temperature_fahrenheit - 58)

discomfort_index = np.where(discomfort_index > 80, 80, discomfort_index)

print("First 10 location's (5 days) Temperature in Fahrenheit: \n", temperature_fahrenheit[:10, :5])
print("First 10 location's (5 days) Discomfort index: \n", discomfort_index[:10, :5])

```

```python
# Task 4
jan_temp = temperature[:, :31]

avg_jan_temp = np.nanmean(jan_temp)  # Calculate mean, ignoring NaNs

print(f"Average January Temperature of all 500 locations: {avg_jan_temp:.2f} °C")

```

```python
#Task 5
temperature_copy = temperature.copy()
temperature_copy[temperature > 35] = np.nan  # Replace temperatures > 35°C with NaN

null_counts_per_loc = np.isnan(temperature_copy).sum(axis=1) # sum all null values

for loc, null_count in enumerate(null_counts_per_loc):
    print(f"Location {loc + 1}: {null_count} potential error values (None)")

```

```python
# Task 6
import numpy as np


days_in_months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Create slices for each quarter
quarter1 = temperature[:, :31]   
quarter2 = temperature[:, 31:121]
quarter3 = temperature[:, 121:273]
quarter4 = temperature[:, 273:]   


avg_quarter_1 = np.nanmean(quarter1, axis=1)
avg_quarter_2 = np.nanmean(quarter2, axis=1)
avg_quarter_3 = np.nanmean(quarter3, axis=1)
avg_quarter_4 = np.nanmean(quarter4, axis=1)

# Combine the averages into a single array
quarterly_averages = np.column_stack((avg_quarter_1,
                                       avg_quarter_2,
                                       avg_quarter_3,
                                       avg_quarter_4))

print("Fist 10 Locations averages quater wise:\n", quarterly_averages[:10])
```

```python
# Task 7
dry_days_count = np.zeros(humidity.shape[0], dtype=int)  # For "Dry" days
humid_days_count = np.zeros(humidity.shape[0], dtype=int)  # For "Humid" days

# Classify humidity levels
dry_days_count = np.sum(humidity < 30, axis=1)  # Count "Dry" days
humid_days_count = np.sum(humidity > 70, axis=1)  # Count "Humid" days

# Display the results
for location in range(humidity.shape[0]):
    print(f"Location {location + 1}: {dry_days_count[location]} Dry days, {humid_days_count[location]} Humid days")

```

```python
# Task 8
mean_pressure = 1013  # Mean atmospheric pressure in hPa as a referenceof adjustment
pressure_variation = np.random.normal(0, 5, (365,))
pressure_trend = mean_pressure + pressure_variation


adjusted_temperature_data = temperature.copy()

for day in range(365):
    pressure_adjustment = (pressure_trend[day] - mean_pressure) * 0.1 
    adjusted_temperature_data[:, day] += pressure_adjustment

print("First 10 Locations with 5 days adjusted temperature: \n", adjusted_temperature_data[:10, :5])
```
